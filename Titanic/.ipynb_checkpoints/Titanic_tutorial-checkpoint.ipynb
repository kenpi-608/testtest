```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
print(test.shape)
print(train.shape)
```

```python
def kesson_table(df):
    null_val = df.isnull().sum()
    percent = 100 * df.isnull().sum() / len(df)
    kesson_table = pd.concat([null_val, percent], axis=1)
    kesson_table_len_columns = kesson_table.rename(columns = {0: '欠損数', 1: '%'})
    return kesson_table_len_columns
```

```python
kesson_table(train)
```

```python
kesson_table(test)
```

```python
train['Age'] = train['Age'].fillna(train['Age'].median())
```

```python
print("S:", np.sum(train['Embarked'] == 'S'))
print("C:", np.sum(train['Embarked'] == 'Q'))
print("Q:", np.sum(train['Embarked'] == 'C'))
```

```python
train['Embarked'] = train['Embarked'].fillna('S')
```

```python
train['Sex'][train['Sex'] == 'male'] = 0
train['Sex'][train['Sex'] == 'female'] = 1
train['Embarked'][train['Embarked'] == "S"] = 0
train['Embarked'][train['Embarked'] == "C"] = 1
train['Embarked'][train['Embarked'] == "Q"] = 2
```

```python
train.head(10)
```

```python
test['Age'] = test['Age'].fillna(test['Age'].median())
test['Sex'][test['Sex'] == 'male'] = 0
test['Sex'][test['Sex'] == 'female'] = 1
test['Embarked'][test['Embarked'] == "S"] = 0
test['Embarked'][test['Embarked'] == "C"] = 1
test['Embarked'][test['Embarked'] == "Q"] = 2
test['Fare'] = test['Fare'].fillna(test['Fare'].median())
```

```python
kesson_table(test)
```

```python
from sklearn import tree
```

```python
target = train["Survived"].values
features = train[["Pclass", "Sex", "Age", "Fare", "SibSp", "Parch", "Embarked"]]
```

```python
d_tree = tree.DecisionTreeClassifier()
d_tree = d_tree.fit(features, target)
```

```python
test_features = test[["Pclass", "Sex", "Age", "Fare", "SibSp", "Parch", "Embarked"]]
```

```python
t_predict = d_tree.predict(test_features)
```

```python
label = pd.read_csv("gender_submission.csv")
```

```python
def accuracy(predict, label):
    return np.sum(predict == label) / len(label)
```

```python
true_label = np.array(label.iloc[:, 1])
```

```python
accuracy(t_predict, true_label)
```

```python
from sklearn.ensemble import RandomForestClassifier as classifier
```

```python
# 木の数
num_tree = 20
r_tree = classifier(n_estimators=num_tree, criterion="gini", oob_score=True, random_state=777)
r_tree.fit(features, target)
```

```python
r_tree_predict = r_tree.predict(test_features)
```

```python
r_tree.feature_importances_
```

```python
accuracy(r_tree_predict, true_label)
```

```python
PassengerId = np.array(test["PassengerId"]).astype(int)
my_answer = pd.DataFrame(r_tree_predict, PassengerId, columns = ["Survived"])
my_answer
```

```python
my_answer.to_csv("my_answer.csv", index_label = ["PassengerId"])
```
