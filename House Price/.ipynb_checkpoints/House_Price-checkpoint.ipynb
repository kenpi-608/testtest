## 欠損値処理

```python
train_XX.columns[train_XX.isnull().sum() > 1]
```

```python
train_XX["MasVnrArea"].fillna(train_XX["MasVnrArea"].median())
train_XX["BsmtQual"].fillna("TA")
train_XX["BsmtCond"].fillna("TA")
train_XX["BsmtExposure"].fillna("No")
train_XX["BsmtFinType1"].fillna("Unf")
train_XX["BsmtFinType2"].fillna("Unf")
train_XX["GarageType"].fillna("Attchd")
train_XX["GarageYrBlt"].fillna(train_XX["GarageYrBlt"].median())
train_XX["GarageFinish"].fillna("Unf")
train_XX["GarageQual"].fillna("TA")
train_XX["GarageCond"].fillna("TA")
```

```python
test_X["MSZoning"].fillna("RA")
test_X["Utilities"].fillna("AllPub")
test_X["MasVnrArea"].fillna(test_X["MasVnrArea"].median())
test_X["BsmtQual"].fillna("TA")
test_X["BsmtCond"].fillna("TA")
test_X["BsmtExposure"].fillna("No")
test_X["BsmtFinType1"].fillna("GLQ")
test_X["BsmtFinType2"].fillna("Unf")
test_X["BsmtFullBath"].fillna(test_X["BsmtFullBath"].median())
test_X["BsmtHalfBath"].fillna(test_X["BsmtHalfBath"].median())
test_X["Functional"].fillna("Typ")
test_X["GarageType"].fillna("Attchd")
test_X["GarageYrBlt"].fillna(test_X["GarageYrBlt"].median())
test_X["GarageFinish"].fillna("Unf")
test_X["GarageQual"].fillna("TA")
test_X["GarageCond"].fillna("TA")
```

```python
unique_list = []
for i in range(len(train_XX.columns)):
    if isinstance(train_XX.iloc[:, i][0], str):
        labels, uniques = pd.factorize(train_XX.iloc[:, i])
        train_XX.iloc[:, i] = labels
        unique_list.append(uniques)
```

```python
t = 0
for i in range(len(test_X.columns)):
    if isinstance(test_X.iloc[:, i][0], str):
        test_X.iloc[:, i] = unique_list[t].get_indexer(test_X.iloc[:, i])
        t = t + 1
```
